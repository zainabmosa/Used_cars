### Data Handling:

```python
df.drop(columns='county',inplace=True)
```

```python
cars= df.copy()
```

```python
categorical_cols = ['condition', 'cylinders', 'fuel', 'title_status','transmission', 'drive', 'size', 'type','paint_color']
cars[categorical_cols] = cars[categorical_cols].fillna('Unknown')
```

```python
cars.shape
```

```python
cars.dropna(subset=['price', 'year', 'odometer', 'manufacturer', 'model'], inplace=True)
```

```python
cars.shape
```

```python
cars = cars[(cars['price'] >= 500) &(cars['price'] <= 200000)]
```

```python
cars.shape
```

```python
cars = cars[(cars['odometer'] >= 10000) &(cars['odometer'] <= 600000)]
```

```python
cars.shape
```

```python
cars['year'] = cars['year'].astype(int)
```

```python
cars = cars[cars.year>=1950]
```

```python
cars.shape
```

## Analysis:

## OBJECTIVE 1
### How can two cars from the same manufacturer and year have different prices?

BMW 2001 Price Comparison

```python
comparison = cars[(cars['manufacturer'] == 'bmw') & (cars['year'] == 2001)][[
    'manufacturer',
    'model',
    'year',
    'price',
    'odometer',
    'condition',
    'transmission',
    'type',
    'size'
]].sort_values('price')

comparison.iloc[[0, -5]]
```

Jeep 2016 Price Comparison

```python
comparison = cars[(cars['manufacturer'] == 'jeep') & (cars['year'] == 2016)&(cars['condition']!='Unknown')][[
    'manufacturer',
    'model',
    'year',
    'price',
    'odometer',
    'condition',
    'transmission',
    'type',
    'size'
]].sort_values('price')

comparison.iloc[[0, -1]]
```

```python
toyota_models = cars[cars['manufacturer'] == 'bmw']

model_price = (toyota_models.groupby('model')['price'].median().sort_values(ascending=False).sample(5))

model_price.plot(kind='bar', figsize=(10,5), color='pink')

plt.xlabel('bmw Models')
plt.ylabel('Median Price')
plt.xticks(rotation=45)

plt.show()
```

```python
toyota_models = cars[cars['manufacturer'] == 'jeep']

model_price = (toyota_models.groupby('model')['price'].median().sort_values(ascending=False).sample(5))

model_price.plot(kind='bar', figsize=(10,5), color='pink')

plt.xlabel('jeep Models')
plt.ylabel('Median Price')
plt.xticks(rotation=45)

plt.show()
```

## OBJECTIVE 3
### How does odometer mileage affect vehicle prices?

```python
cars['Mileage Range'] = pd.cut(cars['odometer'],bins=[10000,60000,100000,150000,200000,300000,600000])

odometer_price = cars.groupby('Mileage Range')['price'].median()

odometer_price.plot(kind='line', marker='o', figsize=(8,5), color='pink')

plt.title('Median Used Car Price by Odometer Mileage')
plt.xlabel('Odometer Mileage')
plt.ylabel('Median Price')
plt.xticks(rotation=45)

plt.show()
```

## OBJECTIVE 4
### How do used car prices vary across different vehicle conditions?

```python
c_p = cars.groupby('condition')['price'].median().sort_values()
c_p = c_p.drop(['Unknown','new'])

c_p.plot(kind='bar', figsize=(8,5), color='pink')

plt.title('Median Used Car Price by Vehicle Condition')
plt.xlabel('Condition')
plt.ylabel('Median Price')
plt.xticks(rotation=45)
plt.show()
```

## OBJECTIVE 5
### How do used car prices vary across different vehicle types?

```python
t_p = cars.groupby('type')['price'].median().sort_values()
t_p = t_p.drop(['Unknown','other'])
t_p.plot(kind='bar', figsize=(10,5), color='pink')

plt.title('Median Used Car Price by Vehicle Type')
plt.xlabel('Vehicle Type')
plt.ylabel('Median Price')
plt.xticks(rotation=45)
plt.show()
```

## OBJECTIVE 6
### How does the manufacturing year affect vehicle prices?

```python
year_price = cars.groupby('year')['price'].median().sort_index()

year_price.plot(kind='line',figsize=(12, 5),marker='o', color='pink')

plt.title('Median Used Car Price by Manufacturing Year')
plt.xlabel('Manufacturing Year')
plt.ylabel('Median Price')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
print('2022:')
print(cars[['year','price']][cars.year==2022].mean())
print('\n2021:')
print(cars[['year','price']][cars.year==2021].mean())
print('\n2020:')
print(cars[['year','price']][cars.year==2020].mean())
```

## OBJECTIVE 7
### How do used car prices vary across different manufacturers?

```python
top10 = cars['manufacturer'].value_counts().head(10).index

m_p = (cars[cars['manufacturer'].isin(top10)].groupby('manufacturer')['price'].median().sort_values(ascending=False))

m_p.plot(kind='bar', figsize=(10,5), color='pink')

plt.title('Top 10 Manufacturers by Median Used Car Price')
plt.xlabel('Manufacturer')
plt.ylabel('Median Price')
plt.xticks(rotation=45)

plt.show()
```

# Summary:

- Even cars from the same manufacturer and year can have different prices because of the model.

- Even cars from the same manufacturer, model, and year can have different prices because of factors such as mileage, condition, transmission, and vehicle type.

- Odometer mileage has a strong effect on price. Cars with lower mileage generally have higher prices.

- Used car prices vary depending on the vehicle condition, and cars in better condition generally have higher prices.

- Vehicle type also has a noticeable effect on used car prices, Especially full size.

- Prices show a slight increase for very old vehicles, remain lower for some middle-year models, then increase significantly for newer vehicles. However, prices decrease slightly for the 2022 model year.

- The manufacturer has a strong effect on used car prices.
