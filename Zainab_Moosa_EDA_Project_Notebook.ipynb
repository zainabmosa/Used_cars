### Data Handling:

```python
df.drop(columns='county',inplace=True)
```

```python
cars= df.copy()
```

```python
categorical_cols = ['condition', 'cylinders', 'fuel', 'title_status','transmission', 'drive', 'size', 'type','paint_color']
cars[categorical_cols] = cars[categorical_cols].fillna('Unknown')
```

```python
cars.shape
```

```python
cars.dropna(subset=['price', 'year', 'odometer', 'manufacturer', 'model'], inplace=True)
```

```python
cars.shape
```

```python
cars = cars[(cars['price'] >= 500) &(cars['price'] <= 100000)]
```

```python
cars.shape
```

```python
cars = cars[(cars['odometer'] >= 0) &(cars['odometer'] <= 500000)]
```

```python
cars.shape
```

```python
cars['year'] = cars['year'].astype(int)
```

```python
cars = cars[cars.year>=1980]
```

## Analysis:

## OBJECTIVE 1
### How can two cars from the same manufacturer and year have different prices?

```python
comparison = cars[(cars['manufacturer'] == 'bmw') & (cars['year'] == 2001)][[
    'manufacturer',
    'model',
    'year',
    'price',
    'odometer',
    'condition',
    'transmission',
    'type',
    'size'
]].sort_values('price')

comparison.iloc[[0, -5]]
```

```python
comparison = cars[(cars['manufacturer'] == 'honda') & (cars['year'] == 2016)][[
    'manufacturer',
    'model',
    'year',
    'price',
    'odometer',
    'condition',
    'transmission',
    'type',
    'size'
]].sort_values('price')

comparison.iloc[[0, -1]]
```

```python
comparison = cars[(cars['manufacturer'] == 'jeep') & (cars['year'] == 2016)][[
    'manufacturer',
    'model',
    'year',
    'price',
    'odometer',
    'condition',
    'transmission',
    'type',
    'size'
]].sort_values('price')

comparison.iloc[[0, -1]]
```

## OBJECTIVE 2
### How do used car prices vary across different vehicle conditions?

```python
c_p = cars.groupby('condition')['price'].median().sort_values()
c_p = c_p.drop('Unknown')

c_p.plot(kind='bar', figsize=(8,5), color='pink')

plt.xlabel('Condition')
plt.ylabel('Median Price')
plt.xticks(rotation=45)
plt.show()
```

## OBJECTIVE 3
### How does odometer mileage affect vehicle prices?

```python
cars['Mileage Range'] = pd.cut(
    cars['odometer'],
    bins=[0,50000,100000,150000,200000,300000,500000]
)

odometer_price = cars.groupby('Mileage Range')['price'].median()

odometer_price.plot(kind='line', marker='o', figsize=(8,5), color='pink')

plt.xlabel('Odometer Mileage')
plt.ylabel('Median Price')
plt.xticks(rotation=45)

plt.show()
```

## OBJECTIVE 4
### How do used car prices vary across different vehicle sizes and models?

```python
size_price = cars[cars['size'] != 'Unknown'].groupby('size')['price'].median()

size_price.plot(kind='bar', figsize=(8,5), color='pink')

plt.xlabel('Vehicle Size')
plt.ylabel('Median Price')
plt.xticks(rotation=0)

plt.show()
```

```python
toyota_models = cars[cars['manufacturer'] == 'toyota']

model_price = (toyota_models.groupby('model')['price'].median().sort_values(ascending=False).sample(5))

model_price.plot(kind='bar', figsize=(10,5), color='pink')

plt.xlabel('Toyota Models')
plt.ylabel('Median Price')
plt.xticks(rotation=45)

plt.show()
```

## OBJECTIVE 5
### How do used car prices vary across different vehicle types?

```python
t_p = cars.groupby('type')['price'].median().sort_values()
t_p = t_p.drop(['Unknown','other'])
t_p.plot(kind='bar', figsize=(10,5), color='pink')

plt.xlabel('Vehicle Type')
plt.ylabel('Median Price')
plt.xticks(rotation=45)
plt.show()
```

## OBJECTIVE 6
### How does the manufacturing year affect vehicle prices?

```python
year_price = cars.groupby('year')['price'].median().sort_index()

year_price.plot(kind='line',figsize=(12, 5),marker='o', color='pink')

plt.xlabel('Manufacturing Year')
plt.ylabel('Median Price')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

## OBJECTIVE 7
### How do used car prices vary across different manufacturers?

```python
top10 = cars['manufacturer'].value_counts().head(10).index

m_p = (
    cars[cars['manufacturer'].isin(top10)]
    .groupby('manufacturer')['price']
    .median()
    .sort_values(ascending=False)
)

m_p.plot(kind='bar', figsize=(10,5), color='pink')

plt.xlabel('Manufacturer')
plt.ylabel('Median Price')
plt.xticks(rotation=45)

plt.show()
```

## OBJECTIVE 8
### How do prices vary between automatic and manual vehicles?

```python
tr_p = cars.groupby('transmission')['price'].median().sort_values()
tr_p = tr_p.drop(['Unknown','other'])

tr_p.plot(kind='bar', figsize=(6,5), color='pink')

plt.xlabel('Transmission')
plt.ylabel('Median Price')
plt.xticks(rotation=0)
plt.show()
```

# Summary:

- Even cars from the same manufacturer and year can have different prices because of factors such as mileage, condition, transmission, and vehicle type.

- Used car prices vary depending on the vehicle condition, and cars in better condition generally have higher prices.

- Odometer mileage has a strong effect on price. Cars with lower mileage generally have higher prices.

- Vehicle size and model affect used car prices. Full-size vehicles generally have higher prices, and some vehicle models are more expensive than others.

- Vehicle type also has a noticeable effect on used car prices.

- Prices show a slight increase for very old vehicles, remain lower for some middle-year models, then increase significantly for newer vehicles. However, prices slightly decrease for the 2022 model year .

- The manufacturer has a strong effect on used car prices.

- Transmission type has some effect on price, but the difference between automatic and manual vehicles is not very large.
